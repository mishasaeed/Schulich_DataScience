```python
import pandas as pd

# Group the crime report data by 'REPORT_YEAR' and count the number of cases in each year
case_by_year = cm.groupby('REPORT_YEAR').size().reset_index(name='total_cases')

# Convert case_by_year.values to a list
tc = pd.DataFrame({'Year': case_by_year['REPORT_YEAR'].values.tolist(), 'total_cases': case_by_year['total_cases'].values.tolist()})

data = {
    'Year': [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022],
    'Population': [5797000, 5867000, 5938000, 6010000, 6082000, 6139000, 6197000, 6255000, 6313000]
}


pop = pd.DataFrame(data)

merged_data = pd.merge(tc, pop, left_on='Year', right_on='Year')

# Calculate the crime rate by dividing the number of cases by the population and multiplying by 100,000
merged_data['Crime_Rate'] = (merged_data['total_cases'] / merged_data['Population']) * 100000

print(merged_data[['Year', 'Crime_Rate']])
```

```python
# Create a line chart
plt.plot(merged_data['Year'], merged_data['Crime_Rate'])

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Crime Rate')
plt.title('Crime Rate by Year')

# Display the chart
plt.show()
```

```python
merge_df3_crimeseverity = pd.merge(df3,csi, on='OCC_YEAR')
merge_df3_crimeseverity
```

```python
unemployment_rate = {'OCC_YEAR':[2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021],
                      'unemployment_rate': [7.19,6.76,6.32,6.04,6.14,8.34,8.06,7.51,7.29,7.07,6.91,6.91,7,6.34,5.83,5.66,9.46,7.51]}
unemployment_df = pd.DataFrame(unemployment_rate)
unemployment_df
```

```python
population_data = {
    'OCC_YEAR': [2004,2005,2006,2007,2008,2009,2010,2011,2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022],
    'Population': [4948000,5035000,5125000,5216000,5309000,5403000,5499000,5593000,5660000,5728000,5797000, 5867000, 5938000, 6010000, 6082000, 6139000, 6197000, 6255000, 6313000]
}
population_df = pd.DataFrame(population_data)
population_df
```

```python
merge_final = pd.merge(merge_df3_crimeseverity,unemployment_df, on='OCC_YEAR')
merge_final = pd.merge(merge_final,population_df, on='OCC_YEAR')
merge_final = merge_final.drop('HOOD_158', axis=1)
merge_final = merge_final.drop('DIVISION', axis=1)
merge_final
```

```python
replacement_map = {'Morning': 1, 'Afternoon': 2, 'Evening': 3,'Night':4}
merge_final['OCC_TIME_RANGE'] = merge_final['OCC_TIME_RANGE'].replace(replacement_map)
```

```python
# separate features and target
categorical_features = ['OCC_TIME_RANGE']
numerical_features = ['CSI','unemployment_rate', 'OCC_YEAR','OCC_DOW','OCC_DAY','OCC_MONTH','OCC_HOUR','OCC_DOY','Population']
target = 'INJURIES'
```

```python
# split data in train / test
X_train, X_test, y_train, y_test = train_test_split(merge_final[categorical_features + numerical_features]
                                                    ,merge_final[target],
                                                    test_size = 0.3, random_state=1234)
```

```python
params = {
    "n_estimators": 100,  # Number of trees in the forest
    "max_depth": 5,  # Max depth of the tree
    "min_samples_split": 4,  # Min number of samples required to split a node
    "min_samples_leaf": 2,  # Min number of samples required at a leaf node
    "ccp_alpha": 0,  # Cost complexity parameter for pruning
    "random_state": 123,
}
```

```python
rf = RandomForestRegressor(**params)
rf.fit(X_train,y_train)
```

```python
pred = rf.predict(X_test)
```

```python
np.sqrt(mean_squared_error(pred,y_test))
```

```python


# Create a series of feature importances
importance_series = pd.Series(data=rf.feature_importances_, index=X_train.columns)

# Plot a horizontal barplot of the feature importances
importance_series.plot(kind="barh")
```
